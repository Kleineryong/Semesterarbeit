```python
import os
import pandas as pd
import numpy as np
import tifffile as tif
import matplotlib.pyplot as plt
from scipy.optimize import Bounds,minimize
from scipy.constants import c, h, k
currentdir = os.getcwd()
homefolder = os.path.dirname(os.path.dirname(currentdir))
```

```python
camera_folder = homefolder+"/13_Databank/"
DF_F = pd.read_excel(camera_folder+"CMS22010236.xlsx",'Filter')
DF_QE = pd.read_excel(camera_folder+"CMS22010236.xlsx",'QE')
DF_T = pd.read_excel(camera_folder+"FIFO-Lens_tr.xlsx")
```

```python
qe_matrix = np.array(DF_QE.iloc[:-1, 1:])
```

```python
print(np.shape(qe_matrix))
```

```python
def GT(temp,wavelength):
    result_value = param1/(wavelength**5)/(np.exp(param2/(wavelength*temp))-1)*10**(-6)
    return result_value
```

```python
qe_array = np.array(DF_QE.iloc[:-1, [0,1]]).transpose()
param1 = h * 2 * c ** 2
param2 = h * c / k
def CT_fit(x, i, wl0, wl1):
    qe_std = qe_matrix[:, i]*GT(1800, qe_array[0]*10**(-9))
    for k in range(9):
        if k != i:
            qe_std = qe_std - x[k] * qe_matrix[:, k] *GT(1800, qe_array[0]*10**(-9))
    CT_target = np.concatenate([qe_std[:wl0], qe_std[wl1:]])
    return np.sum(np.abs(CT_target))

# def CT_fit(x, i, wl0, wl1):
#     qe_std = qe_matrix[:, i]
#     for k in range(9):
#         if k != i:
#             qe_std = qe_std - x[k] * qe_matrix[:, k]
#     CT_target = np.concatenate([qe_std[:wl0], qe_std[wl1:]])
#     return np.sum(np.abs(CT_target))

x0 = -np.ones(9)
bounds = Bounds([-1] * 9, [1] * 9)
result = minimize(CT_fit, x0, args=(0, 0, 36), bounds=bounds)
```

```python
wl0_array = [0,0,22,35,49,54,68,78]
wl1_array = [36,52,62,74,85,93,107,136]
```

```python
x_matrix = []
for i in range(8):
    result = minimize(CT_fit,x0,args=(i,wl0_array[i],wl1_array[i]),bounds=bounds)
    x_matrix.append(result.x)
```

```python
qe_cor =  np.array(qe_matrix) @ (-np.array(x_matrix).T)
df = pd.DataFrame(qe_cor)
df.to_excel("H:/1_Forschung/12_Program_p\MSI_preprocessing\Crosstalk/qe_cor.xlsx", index=False)
df.to_excel("H:/1_Forschung/13_Databank/qe_cor.xlsx", index=False)
df1 = pd.DataFrame(-np.array(x_matrix).T)
df1.to_excel("H:/1_Forschung/12_Program_p\MSI_preprocessing\Crosstalk/CT_cor.xlsx", index=False)
df1.to_excel("H:/1_Forschung/13_Databank/CT_cor.xlsx", index=False)
for i in range(8):
    plt.clf()
    plt.plot(qe_cor[:,i], label="after_ch_"+str(i+1))
    plt.plot(qe_matrix[:,i],label="before_ch_"+str(i+1))
    plt.legend()
    plt.grid()
    plt.savefig("H:/1_Forschung/12_Program_p\MSI_preprocessing\Crosstalk/compare_"+str(i+1)+".png", dpi=800,bbox_inches='tight')
```

```python
colorset = ['red', 'blue', 'green', 'orange', 'purple', 'teal', 'cyan', 'magenta']
p_T = 1000
for i in range(8):
    plt.clf()
    # your data
    qe_array = np.array(DF_QE.iloc[:-1, [0,1+i]]).transpose()
    qe_array2 = qe_cor[:, i]
    max_value1 = np.max(qe_array2[:])  # Maximum y-value for normalization
    max_value2 = np.max(GT(p_T, qe_array[0]*10**(-9)))
    max_value3 = np.max(10**(6)*qe_array2[:]*GT(p_T, qe_array[0]*10**(-9)))
    lns1 = plt.plot(qe_array[0], qe_array2[:]/max_value1, '--', label='QE_ch'+str(i+1), color=colorset[i])
    lns2 = plt.plot(qe_array[0], GT(p_T, qe_array[0]*10**(-9))/max_value2, label='Plank at '+str(p_T)+"K", color='brown')
    lns3 = plt.plot(qe_array[0], 10**(6)*qe_array2[:]*GT(p_T, qe_array[0]*10**(-9))/max_value3, label='QE$\cdot$PR', color='k')
    plt.grid()
    # set labels
    plt.xlabel("wavelength(µm)")
    plt.ylabel("normalized value")

    plt.legend(loc='upper left')  # Adjust the legend position if needed
    plt.savefig("H:/1_Forschung/12_Program_p\MSI_preprocessing\Crosstalk/re"+str(i+1)+".png", dpi=400, bbox_inches='tight')
    plt.show()
```
